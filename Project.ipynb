```python
# merging the "TESLA" datasets on same date wise as the market will be closed on weekends 
df1=pd.read_csv(r"C:\Users\Jaya Teja\Finance_Project\News Final data sets\tesla_final.csv") # tesla news dataset 
df2=pd.read_csv(r"C:\Users\Jaya Teja\Finance_Project\Price Final data sets\TSLA.csv") # tesla price dataset
```

```python
df1['Date']=pd.to_datetime(df1['Date'],dayfirst=True)
```

```python
tsla_merged=pd.merge(df2,df1,on='Date',how='left')
```

```python
tsla_merged['Label']=0
tsla_merged.loc[0,'Label']=1
```

```python
tsla_merged
```

```python
for i in range(1,len(tsla_merged)):
    if tsla_merged.loc[i,'Adj Close']>= tsla_merged.loc[i-1,'Adj Close']:
        tsla_merged.loc[i,'Label']=1
    else :
        tsla_merged.loc[i,'Label']=0
```
